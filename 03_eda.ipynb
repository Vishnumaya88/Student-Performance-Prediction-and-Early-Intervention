Loading Dataset

```python
import pandas as pd

df = pd.read_csv("student-mat.csv", sep=';')
```

```python
df['pass'] = df['G3'].apply(lambda x: 1 if x >= 10 else 0)
```

Distribution of Pass and Fail Students

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(5,4))
sns.countplot(x='pass', data=df)
plt.title("Pass vs Fail Distribution")
plt.show()
```

 Study Time vs Pass/Fail

```python
plt.figure(figsize=(6,4))
sns.countplot(x='studytime', hue='pass', data=df)
plt.title("Study Time vs Pass/Fail")
plt.show()
```

Absences vs Pass/Fail

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='pass', y='absences', data=df)
plt.title("Absences vs Pass/Fail")
plt.show()
```

Correlation Heatmap

```python
df_heatmap = df.drop(['G1', 'G2', 'G3'], axis=1)

plt.figure(figsize=(8,6))
sns.heatmap(df_heatmap.select_dtypes(include='number').corr(), cmap='coolwarm')
plt.title("Feature Correlation Heatmap")
plt.show()
```
